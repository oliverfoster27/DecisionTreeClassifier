```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def gini_class(series):
    return 1 - (series[series==1].count() / len(series)) ** 2 - (series[series==0].count() / len(series)) ** 2 
```

```python
def gini_float(df, feat, target):
    value_set = list(set(df[feat]))
    value_set.sort()
    thresh_set = [(x + y)/2 for x, y in zip(value_set[1:], value_set)]
    gini_float = 1
    for thresh in thresh_set:
        gini = gini_class(df.query(f"{feat} > {thresh}")[target])
        if gini_float > gini: 
            gini_float, best_thresh = gini, thresh
    return gini_float, best_thresh
```

```python
df = pd.read_csv(r"./data/Credit.csv", index_col=0)
df['High Balance'] = df['Balance'].apply(lambda x: 1 if x > 800 else 0)
df = df.drop(['Balance', 'Ethnicity'], axis=1)
df['Student'] = df['Student'].apply(lambda x: 1 if x == 'Yes' else 0)
df['Married'] = df['Married'].apply(lambda x: 1 if x == 'Yes' else 0)
df['Gender'] = df['Gender'].apply(lambda x: 1 if x.strip() == 'Male' else 0)

df.head()
```

```python
def fit(df, target):
    
    feature_cols = df.columns.tolist()
    feature_cols.remove(target)
    
    def solver(target, feature_cols, query_strs=[], tree_cache=[]):
        
        # print(tree_cache)

        flat_qs = " & ".join(query_strs)
        if len(query_strs) > 0:
            df_tmp = df.query(flat_qs)
        else:
            df_tmp = df
        
        sum_space = df_tmp[target].sum()
        print(sum_space, len(df_tmp))
        if sum_space == len(df_tmp) or sum_space == 0 or len(feature_cols) == 0:
            # leaf node
            return tree_cache + [(flat_qs, df_tmp[target].max())]

        min_gini = 1
        for feature in feature_cols:

            thresh = None
            if df_tmp[feature].dtype=='int64' and len(set(df_tmp[feature])) <= 2:
                # TODO: Class Feature
                gini = gini_class(df_tmp.query(f"{feature} == 1")[target])
            elif df_tmp[feature].dtype in ['int64', 'float64']:
                # TODO: Float Feature
                gini, thresh = gini_float(df_tmp, feature, target)
            else:
                raise ValueError("Unsupport data type")

            print(gini)
            
            if gini < min_gini:
                min_gini = gini
                split_feature = feature
                split_thresh = thresh

        if split_thresh:
            upside_query = query_strs + [f"({split_feature} > {split_thresh})"]
            downside_query = query_strs + [f"({split_feature} <= {split_thresh})"]
        else:
            upside_query = query_strs + [f"({split_feature} == 1)"]
            downside_query = query_strs + [f"({split_feature} == 0)"]
        
        feature_cols.remove(split_feature)
        tree_cache = solver(target, feature_cols, upside_query, tree_cache)
        tree_cache = solver(target, feature_cols, downside_query, tree_cache)

        return tree_cache

    return solver(target, feature_cols)
```

```python
fit(df, 'High Balance')
```
