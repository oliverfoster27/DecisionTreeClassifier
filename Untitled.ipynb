```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def gini_class(series):
    return 1 - (series[series==1].count() / len(series)) ** 2 - (series[series==0].count() / len(series)) ** 2 
```

```python
def gini_float(df, feat, target):
    value_set = list(set(df[feat]))
    value_set.sort()
    thresh_set = [(x + y)/2 for x, y in zip(value_set[1:], value_set)]
    gini_float = 1
    for thresh in thresh_set:
        gini = gini_class(df.query(f"{feat} > {thresh}")[target])
        if gini_float > gini: 
            gini_float, best_thresh = gini, thresh
    return gini_float, best_thresh
```

```python
df = pd.read_csv(r"./data/Credit.csv", index_col=0)
df['High Balance'] = df['Balance'].apply(lambda x: 1 if x > 800 else 0)
df = df.drop(['Balance', 'Ethnicity'], axis=1)
df['Student'] = df['Student'].apply(lambda x: 1 if x == 'Yes' else 0)
df['Married'] = df['Married'].apply(lambda x: 1 if x == 'Yes' else 0)
df['Gender'] = df['Gender'].apply(lambda x: 1 if x.strip() == 'Male' else 0)

df.head()
```

```python
def solver(df, target, query_strs=[], tree_cache=[]):
    
    print(query_strs)
        
    sum_space = df[target].sum()
    print(sum_space, len(df))
    if sum_space == len(df) or sum_space == 0:
        # leaf node
        return tree_cache + [(" & ".join(query_strs), df[target].max())]
        
    min_gini = 1
    for feature in df.drop([target], axis=1).columns:
        
        thresh = None
        if df[feature].dtype=='int64' and len(set(df[feature])) <= 2:
            # TODO: Class Feature
            gini = gini_class(df.query(f"{feature} == 1")[target])
        elif df[feature].dtype in ['int64', 'float64']:
            # TODO: Float Feature
            gini, thresh = gini_float(df, feature, target)
        else:
            raise ValueError("Unsupport data type")
            
        if gini < min_gini:
            min_gini = gini
            split_feature = feature
            split_thresh = thresh
            
    if split_thresh:
        upside_query = query_strs + [f"({split_feature} > {split_thresh})"]
        downside_query = query_strs + [f"({split_feature} <= {split_thresh})"]
    else:
        upside_query = query_strs + [f"({split_feature} == 1)"]
        downside_query = query_strs + [f"({split_feature} == 0)"]
        
    tree_cache = solver(df.query(" & ".join(upside_query)).drop([split_feature], axis=1), target, upside_query, tree_cache)
    tree_cache = solver(df.query(" & ".join(downside_query)), target, downside_query, tree_cache)
    
    return tree_cache
```

```python
solver(df, 'High Balance')
```
